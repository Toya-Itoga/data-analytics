### 目的
ピタゴラス勝率と重回帰分析を用いた勝率の推定および比較




### 分析

#### ライブラリ・データ読み込み

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
df = pd.read_csv('/content/sample_data/team_batting_w_l_2019-2023.csv')
df.head(10)
```

```python
df.shape
df = df.dropna()
df.shape
```

#### ピタゴラス勝率

```python
Pythagoreas_win_rate = (df['R'] ** 2) / (df['R'] ** 2 + df['RA'] ** 2)
Pythagoreas_win_rate.head()
```

#### 重回帰分析

```python
X = df[['R', 'RA']]
y = df['W-L%']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')

model = LR()
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
```

#### 定量評価

```python
mse_Pythagoreas_win_rate = MSE(df['W-L%'], Pythagoreas_win_rate)
rmse_Pythagoreas_win_rate = np.sqrt(mse_Pythagoreas_win_rate)

mse_linear = MSE(y_test, y_pred_test)
rmse_linear = np.sqrt(mse_linear)

print(f"ピタゴラス勝率RMSE: {rmse_Pythagoreas_win_rate}")
print(f"重回帰分析RMSE: {rmse_linear}")
```

### 結論
ピタゴラス勝率と重回帰分析、両者のRMSEを比較するとピタゴラス勝率の方が値が小さいため正確であると言える
